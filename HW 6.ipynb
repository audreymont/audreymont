```python
#this works for T: (5,500)
import math
from matplotlib import pyplot as plt
import numpy as np

def f(x):
    return(((x**4)*(math.exp(x)))/((math.exp(x)-1.0)**2))



Temp = []
CVlist = []
for T in range (5,500):
    V = 0.001 #cubic centimeters
    p = 6.022*(10**28) #1/cubic centimeters
    D = 428 #Kelvin

    N = 1000
    b = D/T
    a = 0
    h = (b-a)/N
    l = 1.3807*(10**(-23))

    s = (0.5*(a)) + (0.5*f(b))
    for k in range (1,N):
        s += f(a+(k*h))
    
    A = (h*s)
    #print(A)
    

    C_v = 9.0*V*p*l*((T/D)**3)*A
    Temp.append(T)
    CVlist.append(C_v)
    print(T, C_v)

print(Temp[0], CVlist[0])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(Temp,CVlist,s=5,c='k',alpha=0.3)
plt.show()
```
