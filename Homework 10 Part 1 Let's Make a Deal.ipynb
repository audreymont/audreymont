```python
#Let's make a deal (original 3 doors)
import numpy as np
import random

swap_win = 0
noswap_win = 0
swap = 0
noswap = 0

for n in range(0,1000):
    #lose = 0
    doors = ['A','B','C']

    winner = random.choice(doors)

    pick = random.choice(doors)
    #print('winner is:', winner)
    #print('pick is:', pick)

    if winner == pick:
        if winner == 'A':
            doors.remove('C')
        if winner == 'B':
            doors.remove('C')
        if winner == 'C':
            doors.remove('B')
    else:
        if pick == 'A' and winner == 'B':
            doors.remove('C')
        if pick == 'A' and winner == 'C':
            doors.remove('B')
        if pick == 'B' and winner == 'A':
            doors.remove('C')
        if pick == 'B' and winner == 'C':
            doors.remove('A')
        if pick == 'C' and winner == 'A':
            doors.remove('B')
        if pick == 'C' and winner == 'B':
            doors.remove('A')
    #print(doors)
    
#switch or keep
    switch_keep_options = ['switch', 'keep']
    switch_keep = random.choice(switch_keep_options)
    #print(switch_keep)
    if switch_keep == 'keep':
        noswap += 1
    else:
        swap += 1

    if switch_keep == 'keep':
        if pick == winner:
            #print("win")
            noswap_win += 1
        #else:
            #print("lose")
            
    else:
        if pick != winner:
            #print('win')
            swap_win += 1
        #else:
            #print('lose')
            #lose += 1
    #print(swap_win)
percentage_swap = swap_win/swap
percentage_noswap = noswap_win/noswap

print('likelihood of winning when you swap is:', percentage_swap)
print('likelihood of winning when you dont swap is:',percentage_noswap)

#win swap/#swap #win no swap/#no swap
```
